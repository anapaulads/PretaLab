```python
#Importando as bibliotecas que irão ser utilizadas no projeto
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

###Leitura, Tratamento e Limpeza dos dados

```python
#Lendo os arquivos
df_educacao = pd.read_csv('PNAD24_EDUCACAO.csv')
df_dicionario = pd.read_excel('dicionario_colunas.xlsx')
```

```python
#visualizando todas as colunas do PNAD_2024
df_educacao.T
```

```python
#visualizando o dicionário de colunas
df_dicionario
```

```python
#Renomeando as colunas do df_educacao de acordo com o dicionário
df_educacao.rename(columns=dict(zip(df_dicionario['nome_coluna_codigo'], df_dicionario['coluna__perguntas_abrev'])), inplace=True, errors='ignore')
```

```python
df_educacao.T
```

```python
#filtro de idade entre 18 - 35 (Geração Y)
df_educacao = df_educacao[(df_educacao['idade'] >= 18) & (df_educacao['idade'] <= 35)]
df_educacao['idade']
```

```python
#Criando uma função para categorizar a idade em faixa-etária (18-23, 24-29, 30-35)
def categorizar_idade(idade):
    if 18 <= idade <= 23:
        return '18-23'
    elif 24 <= idade <= 29:
        return '24-29'
    elif 30 <= idade <= 35:
        return '30-35'
```

```python
# Aplicando a função para criar a nova coluna com a faixa-etária
df_educacao['faixa_etaria'] = df_educacao['idade'].apply(categorizar_idade)
df_educacao[['idade', 'faixa_etaria']] #verificando a aplicação
```

```python
df_educacao
```

```python
#tratando os valores das colunas conforme o dicionario de dados
df_educacao['raca'] = df_educacao['raca'].map({1: 'Branca', 2: 'Preta', 3: 'Amarela', 4: 'Parda', 5: 'Indígena', 9:'Ignorado'})
df_educacao['sabe_ler_escrever'] = df_educacao['sabe_ler_escrever'].map({1: 'Sim', 2: 'Não'})
```

###Análise

Qual volume de pessoas que sabem ler e escrever?

```python
contagem = df_educacao['sabe_ler_escrever'].value_counts()
contagem
```

```python
#Criando um gráfico de barras
plt.figure(figsize=(8, 5))
ax = contagem.plot(kind='bar', color=['green', 'red'])
for p in ax.patches:  #adicionando os valores exatos em cima das barras
    ax.annotate(int(p.get_height()), (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom')
plt.title('Quantidade de pessoas que sabem ler e escrever')
plt.xlabel('Sabe ler e escrever')
plt.ylabel('Quantidade')
plt.grid(axis='y', linestyle='--') #adiciona uma grade ao gráfico para melhor leitura
plt.show()
```

```python
#realizando a contagem da distribuição no Brasil da Gen Y
df_distribuicao_br = df_educacao.groupby(['sexo', 'raca', 'faixa_etaria', 'sigla_uf', 'sabe_ler_escrever']).size().reset_index(name='contagem')
df_distribuicao_br
```

```python
#criando um df com o filtro de quem sabe ler e escrever
df_distribuicao_sabe_ler = df_distribuicao_br[df_distribuicao_br['sabe_ler_escrever'] =='Sim']
df_distribuicao_sabe_ler['sabe_ler_escrever'].value_counts()
```

```python
# Filtrando o DataFrame
df_filtrado = df_distribuicao_br[
    ((df_distribuicao_br['faixa_etaria'] == '18-23') &
     ((df_distribuicao_br['sexo'] == 'Homem') & (df_distribuicao_br['raca'] == 'Branca') |
      (df_distribuicao_br['sexo'] == 'Mulher') & (df_distribuicao_br['raca'] == 'Preta')))]
df_filtrado
```

```python
# Agrupando por sigla de UF e contando
df_agregado = df_filtrado.groupby(['sigla_uf', 'sexo', 'raca'])['contagem'].sum().reset_index()
df_agregado.head(6)
```

```python
# Pivotando para que cada sexo/raca seja uma coluna
df_pivot = df_agregado.pivot_table(index='sigla_uf', columns=['sexo', 'raca'], values='contagem', fill_value=0)
df_pivot.head(6)
```

```python
# Criando um gráfico de barras horizontais
df_pivot.plot(kind='barh', stacked=False, color=['skyblue', 'violet'], figsize=(10, 6))
plt.title('Homens Brancos e Mulheres Negras (18-23 anos) que Sabem Ler e Escrever')
plt.xlabel('Quantidade')
plt.ylabel('Sigla dos Estados')
plt.legend(title='Sexo e Raça')
plt.grid(axis='x', linestyle='--')
plt.show()
```

```python
#Calculando o total por estado para saber a porcentagem de diferença entre os sexos e raça
df_pivot['total_homens_brancos'] = df_pivot[('Homem', 'Branca')]
df_pivot['total_mulheres_negras'] = df_pivot[('Mulher', 'Preta')]
df_pivot['total'] = df_pivot[['total_homens_brancos', 'total_mulheres_negras']].sum(axis=1)
df_pivot['porcentagem_homens_brancos'] = (df_pivot['total_homens_brancos'] / df_pivot['total']) * 100
df_pivot['porcentagem_mulheres_negras'] = (df_pivot['total_mulheres_negras'] / df_pivot['total']) * 100
df_percentual = df_pivot['diferenca_percentual'] = abs(df_pivot['porcentagem_homens_brancos'] - df_pivot['porcentagem_mulheres_negras'])

df_percentual = df_pivot[['porcentagem_homens_brancos', 'porcentagem_mulheres_negras', 'diferenca_percentual']]
df_percentual
```

```python
#Agrupando e verificando o percentual por Região
df_agregado['regiao'] = np.where(df_agregado['sigla_uf'].isin(['AC', 'AP', 'AM', 'PA', 'RO', 'RR', 'TO']), 'Norte',
                         np.where(df_agregado['sigla_uf'].isin(['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE']), 'Nordeste',
                         np.where(df_agregado['sigla_uf'].isin(['DF', 'GO', 'MS', 'MT']), 'Centro-Oeste',
                         np.where(df_agregado['sigla_uf'].isin(['ES', 'MG', 'RJ', 'SP']), 'Sudeste',
                         np.where(df_agregado['sigla_uf'].isin(['PR', 'RS', 'SC']), 'Sul', 'Outras')))))

df_regiao = df_agregado.groupby(['regiao', 'sexo', 'raca'])['contagem'].sum().reset_index()
df_pivot_regiao = df_regiao.pivot_table(index='regiao', columns=['sexo', 'raca'], values='contagem', fill_value=0)
df_pivot_regiao['total_homens_brancos'] = df_pivot_regiao[('Homem', 'Branca')]
df_pivot_regiao['total_mulheres_negras'] = df_pivot_regiao[('Mulher', 'Preta')]
df_pivot_regiao['total'] = df_pivot_regiao[['total_homens_brancos', 'total_mulheres_negras']].sum(axis=1)
df_pivot_regiao['porcentagem_homens_brancos'] = (df_pivot_regiao['total_homens_brancos'] / df_pivot_regiao['total']) * 100
df_pivot_regiao['porcentagem_mulheres_negras'] = (df_pivot_regiao['total_mulheres_negras'] / df_pivot_regiao['total']) * 100
df_pivot_regiao['diferenca_percentual'] = abs(df_pivot_regiao['porcentagem_homens_brancos'] - df_pivot_regiao['porcentagem_mulheres_negras'])

df_regiao_percentual = df_pivot_regiao[['porcentagem_homens_brancos', 'porcentagem_mulheres_negras', 'diferenca_percentual']]
df_regiao_percentual
```

##**Conclusão da análise**

A análise dos dados da PNAD 2024 revela uma discrepância gigante entre a porcentagem de homens brancos e mulheres negras que sabem ler e escrever, refletindo desigualdades históricas e estruturais no Brasil.

A tabela demonstra que, em todas as unidades federativas e regiões, a porcentagem de homens brancos com essas habilidades é consistentemente maior em comparação com mulheres negras. A diferença percentual é especialmente alarmante em estados como Santa Catarina, onde a taxa de homens brancos é superior a 96%, enquanto a de mulheres negras é apenas 3,8%.

Essa disparidade também se reflete nas regiões. No Sul, por exemplo, a diferença percentual entre homens brancos e mulheres negras chega a 88%, enquanto no Norte e Nordeste, as diferenças também permanecem significativas, embora um pouco menores.

Esses números evidenciam como questões raciais e de gênero afetam profundamente o acesso à educação básica e a alfabetização. Mulheres negras historicamente enfrentam múltiplas camadas de discriminação e exclusão, tanto no acesso à educação quanto no mercado de trabalho, o que perpetua essa desigualdade de oportunidades.

Essa análise destaca a necessidade urgente de políticas públicas que promovam a equidade no acesso à educação para minorias, especialmente para mulheres negras, de modo a diminuir essas disparidades alarmantes e proporcionar um futuro mais justo.
